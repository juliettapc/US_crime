```python
import numpy as np
import pandas as pd
```

```python
cbsa2msa = pd.read_csv('../../Data/crosswalks/cbsa_msa_crosswalk.csv',
                      dtype = {'cbsa': object, 'msa': object, 'fipst': object, 'fipscounty':object, 'ssacounty':object})
```

```python
cbsa2msa = cbsa2msa.loc[:, [c for c in cbsa2msa.columns if 'y20' not in c]]
cbsa2msa
```

```python
fips = pd.read_csv('../../Data/crosswalks/ucr_fips_msa.csv', header=None, 
                   names= ['state', 'stategips', 'ucrfips', 'countyname', 'zoning'],
                   dtype = {'ucrfips': object, 'stategips': object})
def county_remover(x):
    return ' '.join(x.split(' ')[:-1]).upper()

fips.countyname = fips.countyname.apply(county_remover)
fips.head()
```

```python
mdf = pd.merge(fips, cbsa2msa, on=['countyname', 'state'])
```

```python
mdf[mdf.countyname == 'CLARK'].loc[:, ['countyname', 'cbsa', 'cbsaname', 'state']]
```

# Now I need to load my data with the cbsa

```python
cbsadf = pd.read_csv('../../Data/crosswalks/cities_gt_50k_cbsa.csv', dtype = {'CBSA': object})
cbsadf.columns = ['City', 'State', 'cbsa']
cbsadf.head()
```

```python
fullmerge = pd.merge(cbsadf, mdf, on='cbsa')
fullmerge.head()
```

```python
coi = ['City', 'State', 'fipst', 'stategips', 'ucrfips', 'fipscounty', 'cbsa', 'msa', 'ssast']

fullmerge.loc[:, coi].to_csv('../../Data/crosswalks/city_msa_crosswalk.csv')
```
