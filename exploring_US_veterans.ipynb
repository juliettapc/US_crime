## Interpolation to add the missing data for different conflicts in 1969, 1990, 2000, ..., both for all US, and by state:



```python
def get_x_y_inter_extrapolation (df, col_x, col_y, kind_interpolation):

    #print (col_y)

    
    
    df_veterans_non_NANs = df[df[col_y].notnull()]   # i remove the rows with nans in the focus conflict column
    
    min_year_data = df_veterans_non_NANs['Year'].min()
    max_year_data = df_veterans_non_NANs['Year'].max()
    
    x = np.array(df_veterans_non_NANs[col_x].tolist())
    y = np.array(df_veterans_non_NANs[col_y].tolist())

        
    
    len_y = len(y)


     # if i only have 3 points, the kind of interpolation needs to be linear, not cubic (otherwise, gives error)
    if len_y < 4:
        kind_interpolation = 'linear'

    f = interpolate.interp1d(x, y, kind=kind_interpolation, fill_value='extrapolate')  # it returns the interpolation function
    

    min_year_interpolate = 1960
    max_year_interpolate = 2017
    
     
    x_interpolate = np.arange(min_year_interpolate,max_year_interpolate+1,1)  
    y_interpolate = f(x_interpolate)   # use interpolation function returned by `interp1d`
   
    
    
    
    ##### i overwrite the values that are outside min-max year of the conflict with NaNs
    for i in range(len(y_interpolate)):
        year = x_interpolate[i]
            #if y_interpolate[i] <0:
        if year >  max_year_data  or   year <  min_year_data  :
                
            y_interpolate[i] = np.nan
    
    
    
    return x_interpolate, y_interpolate
```
